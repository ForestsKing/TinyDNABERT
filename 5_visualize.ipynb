```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("./result.csv")
df
```

```python
plt.figure(figsize=(9, 3), dpi=500)

plt.grid(True, axis='x', linestyle='-', color='#DADADA', zorder=0)
plt.grid(True, axis='y', linestyle='--', color='#DADADA', zorder=0)

plt.scatter(df["Task"].values, df["SOTA"].values, label="SOTA", zorder=100, c="#2B9F2C", marker="^", s=40)
plt.scatter(df["Task"].values, df["Scratch"].values, label="Scratch", zorder=100, c="#1F77B4", marker="s", s=40)
plt.scatter(df["Task"].values, df["Our"].values, label="Our", zorder=100, c="#FF7F0C", marker="*", s=40)

plt.xlim(-0.5, 17.5)
plt.ylim(-0.05, 1.05)
plt.xticks(rotation=30, ha='right', size=8)
plt.yticks(np.linspace(0, 1, 6), size=8)

plt.legend(loc=2, handlelength=0.5, labelspacing=0.5, fontsize=8)
```
